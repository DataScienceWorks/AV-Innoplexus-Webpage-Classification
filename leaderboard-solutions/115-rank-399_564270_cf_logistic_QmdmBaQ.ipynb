# Now checking the data frame and defining funct. for classification:

```python
df.head()
```

```python
####################Adding the probability columns to the data frame
p=df1['probnews']=np.array(pronews[:,1]).copy()
```

```python
#####################Adding probability of clinical
q=df1['probclinical']=np.array(probclinical[:,1]).copy()
```

```python
###############################   publication
r=df1['probpublication']=np.array(propublication[:,1]).copy()
```

```python
###############################    profile
s=df1['probprofile']=np.array(proprofile[:,1]).copy()
```

```python
###############################   conferences
t=df1['probconferences']=np.array(proconferences[:,1]).copy()
```

```python
##############################    forum
u=df1['probforum']=np.array(proforum[:,1]).copy()
```

```python
##############################   thesis
v=df1['probthesis']=np.array(prothesis[:,1]).copy()
```

```python
df.head()
df.shape
df1['final']=df1['Webpage_id'].copy()
df.shape
```

# Final classification function

```python
for i in range(0,25787):
    if (p[i]>0.5 and p[i]>q[i] and p[i]>r[i] and p[i]>s[i] and p[i]>t[i] and p[i]>u[i] and p[i]>v[i]):
        df1['final'][i]='news'
    elif (q[i]>0.5 and q[i]>p[i] and q[i]>r[i] and q[i]>s[i] and q[i]>t[i] and q[i]>u[i] and q[i]>v[i]):
        df1['final'][i]='clinicaltrials'
    elif (r[i]>0.5 and r[i]>p[i] and r[i]>q[i] and r[i]>s[i] and r[i]>t[i] and r[i]>u[i] and r[i]>v[i]):
        df1['final'][i]='publication'
    elif (s[i]>0.5 and s[i]>p[i] and s[i]>q[i] and s[i]>r[i] and s[i]>t[i] and s[i]>u[i] and s[i]>v[i]):
        df1['final'][i]='profile'
    elif (t[i]>0.5 and t[i]>p[i] and t[i]>q[i] and t[i]>r[i] and t[i]>s[i] and s[i]>u[i] and s[i]>v[i]):
        df1['final'][i]='conferences'
    elif (u[i]>0.5 and u[i]>p[i] and u[i]>q[i] and u[i]>r[i] and u[i]>s[i] and u[i]>t[i] and u[i]>v[i]):
        df1['final'][i]='forum'
    elif (v[i]>0.5 and v[i]>p[i] and v[i]>q[i] and v[i]>r[i] and v[i]>s[i] and v[i]>t[i] and v[i]>u[i]):
        df1['final'][i]='thesis'
    else:
        df1['final'][i]='others'
        
    
```
